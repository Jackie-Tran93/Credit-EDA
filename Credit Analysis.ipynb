```python
# remove outlier
upper_app_CNT_CHILDREN = Q3_app_CNT_CHILDREN + 1.5 * IQR_app_CNT_CHILDREN
app = app[app['CNT_CHILDREN'] < upper_app_CNT_CHILDREN]
sns.boxplot(y = app['CNT_CHILDREN'])
```

### Catogorical Vs Catogorical Variables

#### Set hypothesis: H0 is the 2 variables to be compared are independent. H1 is the 2 variables are dependent.

#### TARGET VS NAME_CONTRACT_TYPE

```python
# I have been using cross tab:
pd.crosstab(app['TARGET'],app['NAME_CONTRACT_TYPE'])
```

```python
pd.crosstab(app['TARGET'],app['NAME_CONTRACT_TYPE']).plot.bar()
```

```python
# Calculating the Chi-Square, I am getting the following results:
stat1, p1, dof1, expected1 = chi2_contingency(pd.crosstab(app['TARGET'],app['NAME_CONTRACT_TYPE']))
alpha = 0.05

print('Significance1=%.5f, p1=%.5f' % (alpha, p1))
if p1 <= alpha:
    print('Dependent (reject H0)')
else:
    print('Indepedent (fail to reject Ho)')
```

#### -> there is dependence between TARGET AND NAME_CONTRACT_TYPE
